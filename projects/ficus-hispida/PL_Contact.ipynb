```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load data from files
data1 = pd.read_csv("./data/26305_PL-Contacts_HBond.dat", delim_whitespace=True)
data2 = pd.read_csv("./data/66728267_PL-Contacts_HBond.dat", delim_whitespace=True)
data3 = pd.read_csv("./data/5281426_PL-Contacts_HBond.dat", delim_whitespace=True)
data4 = pd.read_csv("./data/3397_PL-Contacts_HBond.dat", delim_whitespace=True)
```

```python
from matplotlib.ticker import MultipleLocator

color1 = 'tab:purple'
color2 = 'tab:green'
color3 = 'tab:orange'
color4 = 'tab:cyan'

fig, ax = plt.subplots(figsize=(10,6))

# Plot the histogram
plt.hist(data1['Frame#'], bins=50, label="Nodakenatin (CID: 26305)", color=color1)
plt.hist(data2['Frame#'], bins=50, label="Isowigtheone (CID: 66728267)", color=color2)
plt.hist(data3['Frame#'], bins=50, label="7-hydroxycoumarin (CID: 5281426)", color=color3)
plt.hist(data4['Frame#'], bins=50, label="Flutamide (CID: 3397)", color="tab:red")

# Set x-axis tick marks to show every 5th number
ax.xaxis.set_major_locator(MultipleLocator(10))


plt.xlabel('Residue Number')
plt.ylabel('Number of Hydrogen Bonds')
plt.title('Hydrogen Bond Histogram')

plt.legend()
plt.savefig('H-Bond_analysis.png', dpi=300, bbox_inches='tight')
plt.show()
```
