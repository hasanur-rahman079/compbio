```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read in the RMSF data file as a pandas DataFrame
data1 = pd.read_csv('./data/26305_P_RMSF.dat', delim_whitespace=True)
data2 = pd.read_csv("./data/66728267_P_RMSF.dat", delim_whitespace=True)
data3 = pd.read_csv("./data/5281426_P_RMSF.dat", delim_whitespace=True)
data4 = pd.read_csv("./data/3397_P_RMSF.dat", delim_whitespace=True)
```

```python
print(data1.columns)
```

```python
data1.head()
```

```python
from matplotlib.ticker import MultipleLocator

color1 = 'tab:purple'
color2 = 'tab:green'
color3 = 'tab:orange'
color4 = 'tab:cyan'

fig, ax = plt.subplots(figsize=(10,6))

data1["CA"].plot(label="Nodakenatin (CID: 26305)", color=color1)
data2["CA"].plot(label="Isowigtheone (CID: 66728267)", color=color2)
data3["CA"].plot(label="7-hydroxycoumarin (CID: 5281426)", color=color3)
data4["CA"].plot(label="Flutamide (CID: 3397)", color="tab:red")

ax.xaxis.set_major_locator(MultipleLocator(10))

plt.margins(0, 0)
plt.xlabel('Residues')
plt.ylabel('RMSF Å')
plt.title('Protein RMSF')

plt.legend()

plt.savefig('Protein_RMSF.png', dpi=300, bbox_inches='tight')
plt.show
```
