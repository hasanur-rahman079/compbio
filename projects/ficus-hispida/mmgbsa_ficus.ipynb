```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# mmgbsa data file as a pandas DataFrame
data1 = pd.read_csv('./data/26305-trajectory-mmgbsa.csv' )
data2 = pd.read_csv("./data/66728267-trajectory-mmgbsa.csv")
data3 = pd.read_csv("./data/5281426-trajectory-mmgbsa.csv")
data4 = pd.read_csv("./data/3397-trajectory-mmgbsa.csv")
```

```python
print(data1.columns)
```

```python
# Renaming columns in data1
new_columns = {}
for column in data1.columns:
    new_columns[column] = column.replace("r_psp_", "")
data1 = data1.rename(columns=new_columns)

# Renaming columns in data2
new_columns = {}
for column in data2.columns:
    new_columns[column] = column.replace("r_psp_", "")
data2 = data2.rename(columns=new_columns)

# Renaming columns in data3
new_columns = {}
for column in data3.columns:
    new_columns[column] = column.replace("r_psp_", "")
data3 = data3.rename(columns=new_columns)


# Renaming columns in data4
new_columns = {}
for column in data4.columns:
    new_columns[column] = column.replace("r_psp_", "")
data4 = data4.rename(columns=new_columns)
```

```python
print(data1.columns)
```

```python
data1[['MMGBSA_dG_Bind', 'MMGBSA_dG_Bind_Coulomb', 'MMGBSA_dG_Bind_Covalent', 'MMGBSA_dG_Bind_Hbond', 'MMGBSA_dG_Bind_Lipo', 'MMGBSA_dG_Bind_Solv_GB', 'MMGBSA_dG_Bind_vdW']].head()
```

```python
# Selecting the columns to plot
columns_to_plot = ['MMGBSA_dG_Bind', 'MMGBSA_dG_Bind_Coulomb', 'MMGBSA_dG_Bind_Covalent', 'MMGBSA_dG_Bind_Hbond', 'MMGBSA_dG_Bind_Lipo', 'MMGBSA_dG_Bind_Solv_GB', 'MMGBSA_dG_Bind_vdW']

# figure
fig, ax = plt.subplots(figsize=(10, 6))

# colors to use for each data file
colors = ['purple', 'green', 'orange', 'r']

# new dataframe with only the selected columns for each data file
data1_to_plot = data1[columns_to_plot]
data2_to_plot = data2[columns_to_plot]
data3_to_plot = data3[columns_to_plot]
data4_to_plot = data4[columns_to_plot]

# the mean value of each column for each data file
mean_values1 = data1_to_plot.mean()
mean_values2 = data2_to_plot.mean()
mean_values3 = data3_to_plot.mean()
mean_values4 = data4_to_plot.mean()

# Plotting the bar charts
bar1 = ax.barh(y=range(len(columns_to_plot)), width=mean_values1, color=colors[0], height=0.2, 
        label='Nodakenatin (CID: 26305)', alpha=0.8, edgecolor='black', lw=1)
bar2 = ax.barh(y=[x + 0.2 for x in range(len(columns_to_plot))], width=mean_values2, color=colors[1], height=0.2, 
        label='Isowigtheone (CID: 66728267)', alpha=0.8, edgecolor='black', lw=1)
bar3 = ax.barh(y=[x + 0.4 for x in range(len(columns_to_plot))], width=mean_values3, color=colors[2], height=0.2, 
        label='7-hydroxycoumarin (CID: 5281426)', alpha=0.8, edgecolor='black', lw=1)
bar4 = ax.barh(y=[x + 0.6 for x in range(len(columns_to_plot))], width=mean_values4, color=colors[3], height=0.2, 
        label='Flutamide (CID: 3397)', alpha=0.8, edgecolor='black', lw=1)

# top bar
for i, bar in enumerate([bar1, bar2, bar3, bar4]):
    for rect in bar:
        width = rect.get_width()
        ax.text(width + 0.01, rect.get_y() + rect.get_height()/2., f'{width:.2f}',
                ha='left', va='center', fontsize=8)

# y-axis tick labels
ax.set_yticks([x + 0.3 for x in range(len(columns_to_plot))])
ax.set_yticklabels(columns_to_plot)

# x-axis label and title of the plot
ax.set_xlabel("Average Energy (kcal/mol)")
ax.set_title("Post Simulation Trajectory MM/GBSA")

# legend for the data files
ax.legend()

# grid lines to separate the columns
ax.grid(axis='y', alpha=0.3)
ax.grid(axis='x', linestyle='--', alpha=0.3)
ax.set_axisbelow(True)

plt.savefig('Post_MMGBSA.png', dpi=300, bbox_inches='tight')
plt.show()
```
