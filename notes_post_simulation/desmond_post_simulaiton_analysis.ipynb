```python
# Load data from files
data1 = np.loadtxt("./RMSD/26350_PL_RMSD.dat")
data2 = np.loadtxt("./RMSD/66728267_PL_RMSD.dat")
data3 = np.loadtxt("./RMSD/5281426_PL_RMSD.dat")

# Convert frame column from ps to ns
# data1[:, 0] = data1[:, 0] / 10
# data2[:, 0] = data2[:, 0] / 10
# data3[:, 0] = data3[:, 0] / 10

# Extract time, Lig_wrt_Protein, and Prot_CA columns from each file
time1, lig_protein1, prot_ca1 = data1[:, 0], data1[:, 5], data1[:, 1]
time2, lig_protein2, prot_ca2 = data2[:, 0], data2[:, 5], data2[:, 1]
time3, lig_protein3, prot_ca3 = data3[:, 0], data3[:, 5], data3[:, 1]

# Convert time from ps to ns
total_time_ns = time[-1] / 1000

# Plot Lig_wrt_Protein vs time for each file
fig, ax1 = plt.subplots(figsize=(10, 6))
# plt.subplots_adjust(top = 1, bottom = 0, right = 1, left = 0, 
#             hspace = 0, wspace = 0)
plt.margins(0, 0)

color1 = 'tab:blue'
color2 = 'tab:green'
color3 = 'tab:orange'

ax1.set_xlabel('Time (ns)')
ax1.set_ylabel('Lig_wrt_Protein (Å)')
ax1.plot(time1, lig_protein1, label="26350", color=color2)
ax1.plot(time2, lig_protein2, label="66728267", color=color3)
ax1.plot(time3, lig_protein3, label="5281426", color=color1)

ax1.tick_params(axis='y'), #labelcolor=color)
# ax1.set_xlim(0, 100)
# ax1.set_xticks(np.arange(0, 101, 20))
ax1.set_title("Lig_wrt_Protein and Prot_CA over Time")

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:red'
ax2.set_ylabel('Prot_CA (Å)', color=color)  # we already handled the x-label with ax1
ax2.plot(time1, prot_ca1, label="Prot_CA", color=color)
ax2.tick_params(axis='y', labelcolor=color)

# Create the legend with both left and right y-axis labels
# lines1, labels1 = ax1.get_legend_handles_labels()
# lines2, labels2 = ax2.get_legend_handles_labels()
# ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper right' )

# Align the legend for the right y-axis labels in flex
fig.tight_layout()  # otherwise the right y-label is slightly clipped
handles1, labels1 = ax1.get_legend_handles_labels()
handles2, labels2 = ax2.get_legend_handles_labels()
handles = handles1 + handles2
labels = labels1 + labels2
ax2.legend(handles, labels, loc="upper left", bbox_to_anchor=(1.05, 1), borderaxespad=0.)


# plt.legend()
# plt.tight_layout()
plt.show()
```
