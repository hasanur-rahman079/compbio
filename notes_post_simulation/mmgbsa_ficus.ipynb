```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
# read in the mmgbsa data file as a pandas DataFrame
data1 = pd.read_csv('26305-mmgbsa.csv' )
data2 = pd.read_csv("66728267-mmgbsa.csv")
# data3 = pd.read_csv("5281426_L-Properties.dat", delim_whitespace=True)
data4 = pd.read_csv("3397-mmgbsa.csv")
```

```python
print(data1.columns)
```

```python
# Rename columns in data1
new_columns = {}
for column in data1.columns:
    new_columns[column] = column.replace("r_psp_", "")
data1 = data1.rename(columns=new_columns)

# Rename columns in data2
new_columns = {}
for column in data2.columns:
    new_columns[column] = column.replace("r_psp_", "")
data2 = data2.rename(columns=new_columns)

# Rename columns in data4
new_columns = {}
for column in data4.columns:
    new_columns[column] = column.replace("r_psp_", "")
data4 = data4.rename(columns=new_columns)
```

```python
print(data1.columns)
```

```python
data1[['MMGBSA_dG_Bind', 'MMGBSA_dG_Bind_Coulomb', 'MMGBSA_dG_Bind_Covalent', 'MMGBSA_dG_Bind_Hbond', 'MMGBSA_dG_Bind_Lipo', 'MMGBSA_dG_Bind_Solv_GB', 'MMGBSA_dG_Bind_vdW']].head()
```

```python
# Select the columns to plot
columns_to_plot = ['MMGBSA_dG_Bind', 'MMGBSA_dG_Bind_Coulomb', 'MMGBSA_dG_Bind_Covalent', 'MMGBSA_dG_Bind_Hbond', 'MMGBSA_dG_Bind_Lipo', 'MMGBSA_dG_Bind_Solv_GB', 'MMGBSA_dG_Bind_vdW']

# Create a new figure
fig, ax = plt.subplots(figsize=(10, 6))

# Define the colors to use for each data file
colors = ['purple', 'green', 'r']

# Plot a bar chart for each data file
for i, data in enumerate([data1, data2, data4]):
    # Create a new dataframe with only the selected columns
    data_to_plot = data[columns_to_plot]

    # Calculate the mean value of each column
    mean_values = data_to_plot.mean()

    # Plot a bar chart with a different color for each data file
    mean_values.plot.bar(ax=ax, color=colors[i], position=i, width=0.14)

# Set the x-axis tick labels
ax.set_xticks([0, 1, 2, 3, 4, 5, 6])
ax.set_xticklabels(columns_to_plot)

# Set the y-axis label and title of the plot
ax.set_ylabel("Average MM/GBSA Energy (kcal/mol)")
ax.set_title("Post Simulation Trajectory MM/GBSA")

# Add a legend for the data files
ax.legend(labels=['Nodakenatin (CID: 26305)', 'Isowigtheone (CID: 66728267)', 'Flutamide (CID: 3397)'], loc='upper left')

# Show top values

# Add grid lines to separate the columns
ax.grid(axis='y')
ax.grid(axis='x', linestyle='--', alpha=0.5)
ax.set_axisbelow(True)

# Display the figure
style.use('ggplot')

plt.show()
```
