```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read in the L-Properties data file as a pandas DataFrame
data1 = pd.read_csv('./Lproperties/26350_L-Properties.dat', delim_whitespace=True )
data2 = pd.read_csv("./Lproperties/66728267_L-Properties.dat", delim_whitespace=True)
data3 = pd.read_csv("./Lproperties/5281426_L-Properties.dat", delim_whitespace=True)
data4 = pd.read_csv("./Lproperties/3397_control_L-Properties.dat", delim_whitespace=True)
```

```python
print(data2.columns)
```

```python
data1.plot()
```

```python
data1.head()
```

```python
color1 = 'tab:purple'
color2 = 'tab:green'
color3 = 'tab:orange'
color4 = 'tab:cyan'

fig, ax = plt.subplots(figsize=(10,6))

plt.plot(data1['Frame'] / 10, data1['MolSA'], label="Nodakenatin (CID: 26305)", color=color1)
plt.plot(data2['Frame'] / 10, data2['MolSA'], label="Isowigtheone (CID: 66728267)", color=color2)
plt.plot(data3['Frame'] / 10, data3['MolSA'], label="7-hydroxycoumarin (CID: 5281426)", color=color3)
plt.plot(data4['Frame'] / 10, data4['MolSA'], label="Flutamide (CID: 3397)", color="tab:red")


plt.margins(0, 0)
plt.xlabel('Time (ns)')
plt.ylabel('MolSA Å²')
plt.title('Ligand Properties (MolSA)')

plt.legend(loc="upper right")

plt.savefig('Ligand_MolSA.png', dpi=300, bbox_inches='tight')
plt.show
```

```python

fig, ax = plt.subplots(figsize=(10,6))

plt.plot(data1['Frame'] / 10, data1['SASA'], label="Nodakenatin (CID: 26305)", color=color1)
plt.plot(data2['Frame'] / 10, data2['SASA'], label="Isowigtheone (CID: 66728267)", color=color2)
plt.plot(data3['Frame'] / 10, data3['SASA'], label="7-hydroxycoumarin (CID: 5281426)", color=color3)
plt.plot(data4['Frame'] / 10, data4['SASA'], label="Flutamide (CID: 3397)", color="tab:red")


plt.margins(0, 0)
plt.xlabel('Time (ns)')
plt.ylabel('SASA Å²')
plt.title('Ligand Propertie (SASA)')

plt.legend(loc="upper right")

plt.savefig('Ligand_SASA.png', dpi=300, bbox_inches='tight')
plt.show
```
