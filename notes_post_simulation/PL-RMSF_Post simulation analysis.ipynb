```python
from matplotlib.ticker import MultipleLocator

color1 = 'tab:purple'
color2 = 'tab:green'
color3 = 'tab:orange'
color4 = 'tab:cyan'

fig, ax = plt.subplots(figsize=(10,6))

data1["CA"].plot(label="Nodakenatin (CID: 26305)", color=color1)
data2["CA"].plot(label="Isowigtheone (CID: 66728267)", color=color2)
data3["CA"].plot(label="7-hydroxycoumarin (CID: 5281426)", color=color3)
data4["CA"].plot(label="Flutamide (CID: 3397)", color="tab:red")

ax.xaxis.set_major_locator(MultipleLocator(10))

plt.margins(0, 0)
plt.xlabel('Residues')
plt.ylabel('RMSF Å')
plt.title('Protein RMSF')

plt.legend()

plt.savefig('Protein_RMSF.png', dpi=300, bbox_inches='tight')
plt.show
```

```python
# Create a sample DataFrame with Yes/No values
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David'],
        'IsStudent': ['Yes', 'No', 'Yes', 'No']}
df = pd.DataFrame(data)

# Convert Yes/No values to numeric values
df['IsStudent'] = df['IsStudent'].map({'Yes': 1, 'No': 0})

# Plot the DataFrame
ax = df.plot(kind='bar', x='Name', y='IsStudent', legend=False)
ax.set_xlabel('Name')
ax.set_ylabel('Is Student')
ax.set_ylim([0, 1])  # Set y-axis limits to 0-1 for clarity
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv('./RMSF/26305_P_RMSF.dat', delim_whitespace=True)

# Filter the data to only include rows where LigandContact is Yes or No
data = data[data['LigandContact'].isin(['Yes', 'No'])]

# Create separate data frames for Yes and No values of LigandContact
data_yes = data[data['LigandContact'] == 'Yes']
data_no = data[data['LigandContact'] == 'No']

# Plot the data
fig, ax = plt.subplots()
ax.scatter(data_yes['Residue'], data_yes['CA'], c='blue', label='LigandContact Yes')
ax.scatter(data_no['Residue'], data_no['CA'], c='red', label='LigandContact No')
ax.legend()

# Set the axis labels and title
ax.set_xlabel('Residue')
ax.set_ylabel('CA RMSF (Å)')
ax.set_title('RMSF plot with LigandContact')

# Show the plot
plt.show()
```

```python
fig, ax1 = plt.subplots() #this seperate the figures and creates multiple axis plot

years = [2018, 2019, 2020, 2021]
first = [1, 2, 1, 2]

ax1.plot(years, first, color="blue")

# add 2nd y axix value
second = [10, 15, 17, 16]

ax2 = ax1.twinx() #twinx means share the same x axis value

ax2.plot(years, second, color="green")

# add third axis values
third=[111, 115, 190, 140]

ax3 = ax1.twinx()
ax3.plot(years, third, color="red")
# ax3.spines['right'].set_position(('outward', 60)) #spines means the axis border
ax3.spines['right'].set_position(('axes', 1.15)) #does the same thing

ax1.set_ylabel("first", color="blue") #set_ylabel is the label name

ax1.tick_params(axis="y", colors="blue")

plt.show()
```
